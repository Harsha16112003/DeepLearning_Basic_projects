```python
lung_df.head()
```

```python
gender_encoder = LabelEncoder()
LUNG_CANCER = LabelEncoder()
```

```python
# labelEncoding the data
lung_df["GENDER"] = gender_encoder.fit_transform(lung_df["GENDER"])
lung_df["LUNG_CANCER"] = LUNG_CANCER.fit_transform(lung_df["LUNG_CANCER"])
```

```python
lung_df.head()
```

```python
lung_df["LUNG_CANCER"].value_counts()
```

```python
# diving the featuers and the target into x,y
```

```python
x = lung_df.drop("LUNG_CANCER",axis=1)
```

```python
y = lung_df["LUNG_CANCER"]
```

```python
# standrdizing the x or featureset
```

```python
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
```

```python
x
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(sampling_strategy='minority')
x,y=smote.fit_resample(x,y)
y.value_counts()
```

```python
# spliting the data into training and testing data
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# importing the model
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=10,algorithm="kd_tree")
```

```python
model.fit(x_train,y_train)
```

```python
predict=model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
predict
```

```python
print(classification_report(y_test,predict))
```

```python
# to ceate the confusionmatrix
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cnn = confusion_matrix(y_test,predict)
```

```python
sns.heatmap(cnn,annot=True,square=True)
plt.xlabel("predicted")
plt.ylabel("actual")
plt.show()
```
